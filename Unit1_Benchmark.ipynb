Comparing BERT, RoBERTa, and BART across different NLP tasks

```python
from transformers import pipeline
```

```python
gen_bert = pipeline(
    "text-generation",
    model="bert-base-uncased"
)

gen_bert("The future of Artificial Intelligence is", max_length=30)
```

```python
gen_roberta = pipeline(
    "text-generation",
    model="roberta-base"
)

gen_roberta("The future of Artificial Intelligence is", max_length=30)
```

```python
gen_bart = pipeline(
    "text-generation",
    model="facebook/bart-base"
)

gen_bart("The future of Artificial Intelligence is", max_length=30)
```

## Experiment 1: Text Generation
- BERT & RoBERTa struggle or fail
- BART produces coherent text

```python
fill_bert = pipeline(
    "fill-mask",
    model="bert-base-uncased"
)

fill_bert("The goal of Generative AI is to [MASK] new content.")
```

```python
fill_roberta = pipeline(
    "fill-mask",
    model="roberta-base"
)

fill_roberta("The goal of Generative AI is to <mask> new content.")
```

```python
fill_bart = pipeline(
    "fill-mask",
    model="facebook/bart-base"
)

fill_bart("The goal of Generative AI is to <mask> new content.")
```

## Experiment 2: Masked Language Modeling
- BERT and RoBERTa perform strongly
- BART is inconsistent or weak

```python
qa_bert = pipeline(
    "question-answering",
    model="bert-base-uncased"
)

qa_bert(
    question="What are the risks?",
    context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
)
```

```python
qa_roberta = pipeline(
    "question-answering",
    model="roberta-base"
)

qa_roberta(
    question="What are the risks?",
    context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
)
```

```python
qa_bart = pipeline(
    "question-answering",
    model="facebook/bart-base"
)

qa_bart(
    question="What are the risks?",
    context="Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
)
```

## Observation Table

| Task | Model | Classification (Success/Failure) | Observation (What actually happened?) | Why did this happen? (Architectural Reason) |
|------|-------|----------------------------------|----------------------------------------|---------------------------------------------|
| **Generation** | BERT | Failure | The model failed or produced incoherent / unusable output when asked to generate text. | BERT is an encoder-only model and is not trained for autoregressive text generation. |
| | RoBERTa | Failure | Similar to BERT, RoBERTa could not generate meaningful text or threw an error. | RoBERTa is also encoder-only and lacks a decoding mechanism for generation. |
| | BART | Success | Generated a coherent and grammatically correct continuation of the prompt. | BART has an encoder–decoder architecture with an autoregressive decoder designed for generation tasks. |
| **Fill-Mask** | BERT | Success | Correctly predicted suitable words such as "generate" or "create". | BERT is trained using Masked Language Modeling (MLM), making it ideal for this task. |
| | RoBERTa | Success | Produced strong and contextually accurate predictions for the masked word. | RoBERTa is optimized for MLM with improved training strategies and larger data. |
| | BART | Partial / Failure | Predictions were inconsistent or the model struggled compared to BERT/RoBERTa. | BART is not primarily trained for MLM; it is optimized for sequence-to-sequence learning instead. |
| **Question Answering** | BERT | Partial | Returned an answer span, but confidence and accuracy were inconsistent. | The base BERT model is not fine-tuned on QA datasets like SQuAD. |
| | RoBERTa | Partial | Sometimes extracted better answer spans than BERT, but still unreliable. | Strong encoder representations, but lack of QA-specific fine-tuning limits performance. |
| | BART | Partial / Failure | Answers were unstable or less precise than expected. | Although BART supports seq2seq tasks, QA requires explicit fine-tuning which the base model lacks. |
