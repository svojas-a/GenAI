```python
from transformers import pipeline , set_seed , GPT2Tokenizer
```

```python
import os
import nltk
```

```python
file_path = "unit 1.txt"
```

```python
try:
  with open(file_path,'r',encoding="utf-8") as f:
    text=f.read()
  print("File loaded successfully!")
except FileNotFoundError:
  print(f"Error: {file_path} not found!")
```

```python
print("--- Data Preview ---")
print(text[:500]+"...")
```

```python
set_seed(42);
```

```python
prompt="Explain the applications of GenAI"
```

```python
fast_generator = pipeline("text-generation", model="distilgpt2")
output_fast = fast_generator(prompt, max_length=50, num_return_sequences=1)
print(output_fast[0]['generated_text'])
```

```python
fast_generator = pipeline("text-generation", model="gpt2")
output_fast = fast_generator(prompt, max_length=50, num_return_sequences=1)
print(output_fast[0]['generated_text'])
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
```

```python
sample_sentence = "Transformers have attention"
```

```python
tokens = tokenizer.tokenize(sample_sentence)
print(f"Tokens: {tokens}")
```

```python
token_ids = tokenizer.convert_tokens_to_ids(tokens)
print(f"Token IDs: {token_ids}")
```

```python
nltk.download('averaged_perceptron_tagger_eng', quiet=True)
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab')
```

```python
pos_tags = nltk.pos_tag(nltk.word_tokenize(sample_sentence))
print(f"POS Tags: {pos_tags}")
```

```python
ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english")
```

```python
snipper = text[:1000]
entities = ner_pipeline(snipper)

print(f"{'Entity':<20} | {'Type':<10} | {'Score':<5}")
print("-" * 45)

for entity in entities:
    if entity['score'] > 0.9:
        print(f"{entity['word']:<20} | {entity['entity']:<10} | {entity['score']:.2f}")
```

```python
transformer_section = """A Transformer is a deep learning architecture designed to process sequential data by relying on a mechanism called self-attention, which allows the model to weigh the importance of different parts of the input simultaneously. Unlike traditional RNNs or LSTMs that process sequences step by step, Transformers handle all tokens in parallel, making them highly efficient and scalable. They use positional encoding to retain information about word order and are built from stacked encoder and decoder blocks consisting of multi-head attention and feed-forward layers. This design enables Transformers to capture long-range dependencies effectively and has made them the foundation of modern models such as BERT, GPT, T5, and LLaMA, powering state-of-the-art performance in NLP, vision, and multimodal tasks."""
```

```python
fast_sum = pipeline("summarization", model="sshleifer/distilbart-cnn-12-6")
res_fast = fast_sum(transformer_section, max_length=60, min_length=30, do_sample=False)
print(res_fast[0]['summary_text'])
```
